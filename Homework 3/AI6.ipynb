```python
df_new = sw_sf2015.sort_values('Salaries')
```

```python
np.percentile(df_new['Salaries'], 40)
```

```python
np.percentile(df_sf2015['Salaries'], 40)
```

```python
comp = df_sf2015[df_sf2015['Salaries']>9000]['Total Compensation']
aver = []
for i in range(10):
    bootstrap_sample = comp.sample(1000, ignore_index=True)
    mean_temp = np.percentile(bootstrap_sample, 50)
    aver.append(mean_temp)
```

```python
np.mean(aver)
```
